```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.DataFrame({
'h': [150, 175, 170, 180, 190],
'w': [50, 60, 70, 180, 90]
})
df
```

```python
mn = df.min()
mx = df.max()
rg = mx - mn

mn, mx, rg
```

```python
ss = df - mn
ss.head()
```

```python
mm = ss / rg
mm.round(2)
```

```python
mm.plot(kind='bar')
plt.title("Min-Max Scaling [0-1]")
plt.xlabel("Index")
plt.ylabel("Min-max value")
plt.show()
```

```python
md = df.median()
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

md, q1, q3, iqr
```
