```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.DataFrame({
  'h': [150, 160, 170, 185, 190],
  'w': [50, 58, 70, 80, 90]
})
df
```

```python
m = df.mean()
s = df.std()

print("Mean values:\n", m)
print("\nStandard Deviation:\n", s)
```

```python
z = (df - m) / s
z.round(2)
```

```python
z.plot(kind='bar')
plt.title("Z scaling Features")
plt.xlabel("Index")
plt.ylabel("Z value")
plt.show()
```
